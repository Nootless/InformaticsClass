```python
# Import modules you may need to complete HW
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
warnings.simplefilter('ignore')
```

```python
# Read in the 'xnors.csv' dataset and check

data = pd.read_csv('xnors.csv')
print(f'Dataframe:\n{data.head()}')
```

```python
# Separate into X,y
X = data.iloc[:,0:2]
Y = data['label']
print(f'Features:\n{X}')
print(f'Label:\n{Y}')
```

```python
# Visualize the data as a scatter plot
# Show the two classes with different color
# Try to percieve what the decision boundary may look like
plt.scatter(X['x1'],X['x2'],c=Y)
plt.title("x1 vs x2")
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
# Yellow is for True
# Purple is for False
```

```python
# Import Gaussian Naive Bayes Classifier
# Import MLPClassifier from sklearn.NeuralNetwork
# Import train_test_split
# Import classification metrics - accuracy and confusion matrix

from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

```

```python
# Split into train and test and initialize the following parameters
# Set shuffle = True
# Test Size = 20%
# Set a Random State

x_train,x_test,y_train,y_test = train_test_split(X,Y,shuffle=True,test_size=.2,random_state=3)
```

```python
# Instantiate Gaussian NB object
# Then fit to build a model
# Use model to predict
# Print the accuracy of the model
gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_r = gnb.predict(x_test)
print(f'Accuracy of model is {accuracy_score(y_test,y_r)*100} %')
```

```python
# Instantiate  MLPClassifier object
# Then fit to build a model
# Use model to predict
# Print the accuracy of the model

mlp = MLPClassifier()
mlp.fit(x_train,y_train)
y_r = mlp.predict(x_test)
print(f'Accuracy of model is {accuracy_score(y_test,y_r)*100}%\n')
```
